# Jacobian example

```python
import numpy as np
from math import sin, cos
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
class F:
    
    def __init__(self):
        pass
    def __call__(self, x, y):
        F1 = x**2 * y
        F2 = 5 * x + sin(y)
        return np.asarray([F1, F2])
    
    def jac(self, x, y):
        return np.asarray([
            [2*x*y, x**2],
            [5,     cos(y)]
        ])
    def jac_det(self, x, y):
        return (2*x*y * cos(y)) - (5*x**2)
```

```python
f = F()
```

```python
f(2,3)
```

```python
f.jac(2,3)
```

```python
X = np.linspace(0, 10, 100)
Y = np.linspace(0, 10, 100)

v = np.zeros([100, 100])
```

```python
for i in range(len(X)):
    for j in range(len(Y)):
        v[i,j] = f.jac_det(X[i], Y[i])
```

```python
plt.imshow(v)
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x, y)
P = np.sin(X+Y)-2*np.cos(X-Y)
Z = np.exp(P)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, linewidth=0)
```
