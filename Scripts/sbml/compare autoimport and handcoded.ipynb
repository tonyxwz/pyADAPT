# The Trehalose model

## Hand-coded model

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pyADAPT.examples.trehalose import Smallbone2011
```

```python
model = Smallbone2011()
```

```python
model
```

```python
x0 = [
        0.09765, 0.10000, 2.67500, 0.05000, 0.02000, 0.70000, 1.28200, 2.52500,
        1.00000, 0.62500, 1.00000, 1.00000, 0.28150, 0.64910, 1.00000,
        100.00000
]
x0[0] = 100
x0[-1] = 0.1
t = np.linspace(0, 10, 1000)
y = model.compute_states(time_points=t, x0=x0)
```

```python
fluxes = np.array(model.flux_trajectory)
```

```python
plt.plot(fluxes)
```

```python
plt.plot(t, y.T)
plt.xlim([0, 0.5])
plt.ylim([0, 4])
```
